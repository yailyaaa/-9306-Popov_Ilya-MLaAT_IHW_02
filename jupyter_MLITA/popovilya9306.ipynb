#### 9. Исходный код программы.

```python
def read():
    global s
    ret = s[0]
    s = s[1:]
    
    return ret

def Lang(ch):
    if ch == '(':
        ch = read()
        ch = Round(ch)
        if ch == ')':
            ch = read()
            ch = Next(ch)
        else: raise ValueError
    elif ch == '[':
        ch = read()
        ch = Next(ch)
        if ch == ']':
            ch = read()
            ch = Next(ch)
        else: raise ValueError
    elif ch == '#': ch = '#'
    else: raise ValueError
    
    return ch

def Next(ch):
    if ch == ')' or ch == ']' or ch == '#': ch = ch #Пропуск
    else: ch = Lang(ch)
    
    return ch

def Round(ch):
    if ch == '(':
        ch = read()
        ch = Round(ch)
        if ch == ')':
            ch = read()
            ch = Round(ch)
    elif ch == ')' or ch == '#':
        ch = ch #Пропуск
    return ch

s = input('Enter your expression: ')
s = s + '#'
try:
    ch = read()
    ch = Lang(ch)
    print('Entered formula is correct')
except ValueError as ve:
    print('Formula is incorrect')
```
